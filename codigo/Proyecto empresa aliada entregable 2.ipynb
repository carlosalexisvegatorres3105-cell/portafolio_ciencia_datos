```python
# Paso 1: Importar librerías
# pandas y numpy para manejar datos
# matplotlib y seaborn para gráficos
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10,6)

# Paso 2: Cargar los datos
dim_product = pd.read_excel(r"C:\Users\cavega\Desktop\proyecto\DIM_PRODUCT (1).xlsx")
dim_segment = pd.read_excel(r"C:\Users\cavega\Desktop\proyecto\DIM_SEGMENT (1).xlsx")
dim_calendar = pd.read_excel(r"C:\Users\cavega\Desktop\proyecto\DIM_CALENDAR (2).xlsx")
fact_sales = pd.read_csv(r"C:\Users\cavega\Desktop\proyecto\FACT_SALES (1).csv")

# Paso 3: Exploración inicial
print("Dim Product:")
print(dim_product.head(), "\n")

print("Dim Segment:")
print(dim_segment.head(), "\n")

print("Dim Calendar:")
print(dim_calendar.head(), "\n")

print("Fact Sales:")
print(fact_sales.head(), "\n")

# Paso 4: Identificar variables clave
# - fact_sales: WEEK, ITEM_CODE, TOTAL_UNIT_SALES, TOTAL_VALUE_SALES, REGION
# - dim_product: ITEM, CATEGORY, FORMAT, ATTR1, ATTR2, ATTR3
# - dim_segment: CATEGORY, SEGMENT
# - dim_calendar: WEEK, YEAR, MONTH, DATE

# Paso 5: Unir tablas para análisis
# ITEM_CODE - ITEM, CATEGORY - CATEGORY, WEEK - WEEK
df = fact_sales.merge(dim_product, left_on="ITEM_CODE", right_on="ITEM", how="left")
df = df.merge(dim_segment, on="CATEGORY", how="left")
df = df.merge(dim_calendar, on="WEEK", how="left")

# Paso 6: Estadísticas básicas con NumPy
ventas_array = df["TOTAL_VALUE_SALES"].values
print("Promedio:", np.mean(ventas_array))
print("Mediana:", np.median(ventas_array))
print("Desv. estándar:", np.std(ventas_array))
print("Percentil 90:", np.percentile(ventas_array, 90))

# Paso 7: Distribución de ventas por categoría
sns.boxplot(data=df, x="CATEGORY", y="TOTAL_VALUE_SALES")
plt.title("Distribución de ventas por categoría")
plt.xticks(rotation=45)
plt.show()

sns.histplot(data=df, x="TOTAL_VALUE_SALES", bins=30, kde=True, color="skyblue")
plt.title("Histograma de ventas")
plt.show()

# Paso 8: Tendencia de ventas en el tiempo
df["DATE"] = pd.to_datetime(df["DATE"])
ventas_mes = df.groupby(df["DATE"].dt.to_period("M"))["TOTAL_VALUE_SALES"].sum()

ventas_mes.plot(kind="line", marker="o", color="green")
plt.title("Tendencia mensual de ventas")
plt.ylabel("Ventas")
plt.show()

# Paso 9: Relación entre variables
sns.scatterplot(data=df, x="TOTAL_UNIT_SALES", y="TOTAL_VALUE_SALES", hue="SEGMENT")
plt.title("Relación entre unidades y valor de ventas por segmento")
plt.show()

# Paso 10: Ventas por región
ventas_region = df.groupby("REGION")["TOTAL_VALUE_SALES"].sum().reset_index()
sns.barplot(data=ventas_region, x="REGION", y="TOTAL_VALUE_SALES", palette="viridis")
plt.title("Ventas por región")
plt.xticks(rotation=90)
plt.show()

# Paso 11: Outliers
sns.boxplot(data=df, x="REGION", y="TOTAL_VALUE_SALES", palette="Set2")
plt.title("Detección de outliers por región")
plt.xticks(rotation=90)
plt.show()

# Paso 12: Guardar visualizaciones
plt.savefig("grafico_categoria.png")
plt.savefig("grafico_histograma.png")
plt.savefig("grafico_tendencia.png")
plt.savefig("grafico_scatter.png")
plt.savefig("grafico_region.png")
plt.savefig("grafico_outliers.png")

# Paso 13: Documentar insights

# CLORO concentra la mayoría de productos y ventas.
# SAFE BLEACH muestra mayor variabilidad en ventas.
# En diciembre se observa un pico de ventas.
# La región Norte concentra más ventas.
# Se detectan outliers en la región Este.

insights = [
    "CLORO concentra la mayoría de productos y ventas.",
    "SAFE BLEACH muestra mayor variabilidad en ventas.",
    "En diciembre se observa un pico de ventas.",
    "La región Norte concentra más ventas.",
    "Se detectan outliers en la región Este."
]

print("\nINSIGHTS OBTENIDOS:")
for i, insight in enumerate(insights, start=1):
    print(f"{i}. {insight}")
```
